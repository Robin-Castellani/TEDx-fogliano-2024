```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from common import set_plot_style, load_font
```

```python
ROOT_FOLDER = Path("/home/robin/projects/TEDx/")

file_csv = ROOT_FOLDER / "data" / "Emissions" / "ghg-emissions.csv"

images_dir = Path(ROOT_FOLDER / "output" / "images" / "emission")
images_dir.mkdir(parents=True, exist_ok=True)

set_plot_style()
load_font()
```

```python
em = pd.read_csv(
    file_csv, skipfooter=2, engine="python", na_values="false", index_col="Settore"
)
em = em.drop(columns="unit")
em.head()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
em.T.plot.area(stacked=True, ax=ax, cmap="gist_rainbow")
ax.set_title("Emissioni totali di gas serra in Italia")
ax.set_xlabel("Anno")
ax.set_ylabel("Mt CO$_2$eq")

plt.show()
fig.savefig(images_dir / "emissions.png")
```
